```python
from math import *
#Semi-Major Axis
a = 10424.1
#Semi-Minor Axis
b = 9579.42
#Eccentricity
ecc = sqrt(1 - (b**2)/(a**2))
print(ecc)
```

```python
#Circumference
def ellipse_circum(acc=5):
    temp = []
    S_a = 2*pi*a
    S_b = 1
    for x in range(2, acc):
        S_n_a = 2/sqrt(pi)
        S_n_b = (ecc**(2*x-2) * x)/(4*x**2 - 8*x + 3)
        S_n_c = (gamma(0.5 + x))/(factorial(x))
        temp.append(-(S_n_a*S_n_b*S_n_c)/2)
    
    print(temp)
    
    return S_a*(S_b + sum(temp))
        
```

```python
n, A, B, E = Symbol('n'),Symbol('A'),Symbol('B'),Symbol('E')
```

```python
#Circumference
def ellipse_circum_sympy(acc=5):
    temp = []
    S_a = 2*pi*A
    S_b = 1
    for x in range(2, acc):
        S_n_a = 2/sqrt(pi)
        S_n_b = (sqrt(1 - (B**2)/(A**2))**(2*x-2) * x)/(4*x**2 - 8*x + 3)
        S_n_c = (gamma(0.5 + x))/(factorial(x))
        temp.append(-(S_n_a*S_n_b*S_n_c)/2)
    
    print(temp)
    
    return S_a*(S_b + sum(temp))
```

```python
#N=3
factor(ellipse_circum_sympy(3))
```
